```python
response = fraud_ele['Class']
fraud_ele.drop(columns = 'Class',axis=1,inplace=True)
fraud_ele.columns
```

```python
from xgboost import XGBClassifier

from sklearn.model_selection import train_test_split
X_train, Y_train, X_test, Y_test = train_test_split(fraud_ele,response, train_size=0.95)
```

```python
X_train.shape, X_test.shape, Y_train.shape ,Y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline

pipe = Pipeline([('normal',MinMaxScaler()),('xgbclassifier',XGBClassifier())])

pipe.fit(X_train,X_test).score(Y_train,Y_test)
```
